# CLIMATE-FUNDS

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("data.xlsx")
df.rename(columns={'Fund Type':'FundType',},inplace=True)
df.head()
```

### Determine the funds pledged and donated by the countries.

```python
df1=df[['Fund','Pledge (USD mn)','Disbursement (USD mn)']]
df1.rename(columns={'Disbursement (USD mn)':'Donated'},inplace=True)
df1
```

### Compare who gave more, and who gave less funds.

```python
X=list(df1.Fund)
Y=list(df1.Donated)

fig,ax=plt.subplots(figsize=(25,5))
plt.xticks(rotation=90)
ax.bar(X,Y,color='red')
plt.xlabel("Fund",fontsize=25)
plt.ylabel("Amount donated:",fontsize=15)
plt.title("Graph showing the maximum and minimum donation",fontsize=25)
```

```python
df2=df1[['Fund','Donated']]
print("\033[1m"+"It is evident from the above graph that:\n")
print("\033[1m"+"1.The Fund who gave the MAXIMUM donation(in USD mn) is:\n\n")
print(df2[df2.Donated==df2.Donated.max()])
print("\n")
print("\033[1m"+"1.The Fund who gave the MINIMUM donation(in USD mn) is:\n\n")
print(df2[df2.Donated==df2.Donated.min()])
```
